```python
# import the library
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='whitegrid')
```

### Importing the dataset

The dataset is about Black Friday as a study of sales trough consumer behaviours.

```python
# The dataset was downloaded from https://www.kaggle.com/mehdidag/black-friday
# Loading the data
df = pd.read_csv('BlackFriday.csv')
```

```python
fig = plt.figure(figsize=(15, 15))
fig.subplots_adjust(top=0.85, wspace=0.3)
#c = "Set2"

ax1 = fig.add_subplot(3,3,1)
bar1 = sns.countplot(x = "Gender", order=["M","F"], palette="Set2", data = df)

ax2 = fig.add_subplot(3,3,2)
bar2 = sns.countplot(x = "City_Category", order=["A","B","C"], palette="Set2", data = df)

ax3 = fig.add_subplot(3,3,3)
bar3 = sns.countplot(x = "Age", order=["0-17","18-25","26-35","36-45","46-50","55+"], palette="Set2", data = df)

ax4 = fig.add_subplot(3,3,4)
bar4 = sns.countplot(x = "Occupation", palette="Set2", data = df)

ax5 = fig.add_subplot(3,3,5)
bar5 = sns.countplot(x = "Stay_In_Current_City_Years", order=["0","1","2","3","4+"], palette="Set2", data = df)

ax6 = fig.add_subplot(3,3,6)
bar6 = sns.countplot(x = "Marital_Status", palette="Set2", data = df)

ax7 = fig.add_subplot(3,3,7)
bar7 = sns.countplot(x = "Product_Category_1", palette="Set2", data = df)
```

As we can see through these charts, men spent more than women in the retail store, the city B sold more than the others, the costumers between 26 and 35 years old were the ones who most spent in the period, the main customers occupations are the 4, 0, 7 and 1. The costumers that are leaving in the same city for one year are the main buyers in such store. The maried costumers spend more than the single ones. And the products 5, 1 and 8 from category 1 are the best sellers.

```python
# There are no duplicates to be removed
duplicateRowsDF = df[df.duplicated()]
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicateRowsDF)
```

### Dealing with missing values

```python
# Checking for missing values by columns
print(df.isnull().sum())
```

```python
## Assigning value zero for the NaN cases
df.fillna(value=0,inplace=True)
print(df.isnull().sum())
```

### Converting data types

```python
# Verifying the type of the fields
df.dtypes
```

```python
## The Product_ID and categories 2 and 3 could be int.
df['Product_Category_2'] = df['Product_Category_2'].astype(int)
df['Product_Category_3'] = df['Product_Category_3'].astype(int)
df.dtypes
```

### Dealing with outliers

Using Violinplot to find outliers based on City category¶

```python
g = sns.catplot(x = "City_Category", 
                y = "Purchase",
                order = ["A", "B", "C"],
                palette = "husl",
                kind = "violin",
                data = df,
                )
```

As we can see in the Violinplot chart, the distribution of purchases between cities is very similar and there is no reason to remove outliers based on that feature.

Using Boxplot to find outliers based on Gender and City Categorie

```python
sns.catplot(x="City_Category",
            y="Purchase",
            hue="Gender",
            hue_order=["M", "F"],
            order=["A", "B", "C"],
            kind="box",
            palette="pastel",
            data=df);
```

```python
# Trying to find outliers in purchases of products of category 1
print(df['Product_Category_1'].value_counts())
print('mean', np.mean(df['Product_Category_1']))
print('std', np.std(df['Product_Category_1']))
plt.hist(df['Product_Category_1'], bins=20)
plt.show()
```

As we can see, there are some products of category one that could be considered outliers, but we don't want to skew the ranking of the product category.
