# Data Wrangling & Cleaning

```python
# import the library
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the dataset

The dataset is about Black Friday as a study of sales trough consumer behaviours.

```python
# The dataset was downloaded from https://www.kaggle.com/mehdidag/black-friday
# Loading the data
df = pd.read_csv('BlackFriday.csv')
```

### Summarizing your data for inspection

```python
# Cheking the structure of the dataframe
df
```

```python
# Getting more info about the dataset
print('### Blackfriday summary ###')
print('\n')
print(df.columns)
print('\n')
print(df.head())
print('\n')
print(df.describe())
print('\n')
print("Shape of dataframe")
df.shape
```

### Fixing column name

There are no column names to fix

### Removing not relevant columns

All the columns appears to be relevant. Even User_ID and Product_ID can be used depending on the type of analysis to be done.

### Removing duplicates

```python
# There are no duplicates to be removed
duplicateRowsDF = df[df.duplicated()]
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicateRowsDF)
```

### Dealing with missing values

```python
# Checking for missing values by columns
print(df.isnull().sum())
```

```python
## Assigning value zero for the NaN cases
df.fillna(value=0,inplace=True)
print(df.isnull().sum())
```

### Converting data types

```python
# Verifying the type of the fields
df.dtypes
```

```python
## The Product_ID and categories 2 and 3 could be int.
df['Product_Category_2'] = df['Product_Category_2'].astype(int)
df['Product_Category_3'] = df['Product_Category_3'].astype(int)
df.dtypes
```

### Dealing with outliers

Aparently there are no outliers to remove

```python
# Trying to find outliers in users
print(df['User_ID'].value_counts())
print('mean', np.mean(df['User_ID']))
print('std', np.std(df['User_ID']))
plt.hist(df['User_ID'], bins=20)
plt.show()
```

```python
# Trying to find outliers in products of category 1
print(df['Product_Category_1'].value_counts())
print('mean', np.mean(df['Product_Category_1']))
print('std', np.std(df['Product_Category_1']))
plt.hist(df['Product_Category_1'], bins=20)
plt.show()
```

```python
# Trying to find outliers in products of category 2
print(df['Product_Category_2'].value_counts())
print('mean', np.mean(df['Product_Category_2']))
print('std', np.std(df['Product_Category_2']))
plt.hist(df['Product_Category_2'], bins=20)
plt.show()
```

```python
# Trying to find outliers in products of category 3
print(df['Product_Category_3'].value_counts())
print('mean', np.mean(df['Product_Category_3']))
print('std', np.std(df['Product_Category_3']))
plt.hist(df['Product_Category_3'], bins=20)
plt.show()
```
