```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df = pd.read_csv('BlackFriday.csv')

fig = plt.figure()
#fig, ax = plt.subplots()
fig, ax = plt.subplots(4, 3, figsize = (14, 14))
df_col = df.columns

for i, col in enumerate(df_col):

    ax = fig.add_subplot(4, 3, i + 1)
    ax = sns.countplot(x = col, color='orange', data = df)
    ax.set_title(col)
    
    #fig.show()
    
fig.tight_layout()

```

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df = pd.read_csv('BlackFriday.csv')

fig = plt.figure(figsize=(20, 20))
fig.subplots_adjust(top=0.85, wspace=0.3)

ax1 = fig.add_subplot(3,3,1)
ax1.set_title("Gender")
bar1 = sns.countplot(x = "Gender", color="orange", data = df)

ax2 = fig.add_subplot(3,3,2)
ax2.set_title("City Category")
bar2 = sns.countplot(x = "City_Category", color="orange", data = df)

ax3 = fig.add_subplot(3,3,3)
ax3.set_title("Age")
bar3 = sns.countplot(x = "Age", color="orange", data = df)

ax4 = fig.add_subplot(3,3,4)
ax4.set_title("Occupation")
bar4 = sns.countplot(x = "Occupation", color="orange", data = df)

ax5 = fig.add_subplot(3,3,5)
ax5.set_title("Stay_In_Current_City_Years")
bar5 = sns.countplot(x = "Stay_In_Current_City_Years", color="orange", data = df)

ax6 = fig.add_subplot(3,3,6)
ax6.set_title("Marital_Status")
bar6 = sns.countplot(x = "Marital_Status", color="orange", data = df)

ax7 = fig.add_subplot(3,3,7)
ax7.set_title("Product Category 1")
bar7 = sns.countplot(x = "Product_Category_1", color="orange", data = df)

```
